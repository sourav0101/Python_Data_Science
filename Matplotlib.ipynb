```python
# Matlab-style Interface 
plt.figure() # Create a plot figure 

plt.subplot(2, 1, 1)
plt.plot(x, np.sin(x))

plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))
```

```python
# re-create the previous plot using the style of plotting 

# First create a grid of plots
# ax will be an array of two Axes objects 

fig,ax = plt.subplots(2)

# Call plot() method on the appropriate object
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
```

```python
%matplotlib inline 
import matplotlib.pyplot as plt 
plt.style.use('seaborn-whitegrid')
import numpy as np 
```

```python
# SimpleForm for creating all Matplotlib Plots 

fig = plt.figure()
ax = plt.axes() 
```

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0,10,1000)
plt.plot(x, np.sin(x),color='#FA4CD5')
plt.plot(x, np.cos(x),color='#FFDD44')
plt.xlim(-1, 11)  # x Axis limits
plt.ylim(-1.5, 1.5); # y Axis limits
```

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted')
```

```python
plt.plot(x, np.sin(x))
```

```python
plt.plot(x, np.sin(x))
plt.title('A Sine Curve')
plt.xlabel('x')
plt.ylabel('sin(x)')
```

```python
plt.plot(x, np.sin(x),'-g', label='sin(x)')
plt.plot(x, np.cos(x),':b', label='cos(x)')

plt.legend() 
```

```python
%matplotlib inline 
import matplotlib.pyplot as plt 
plt.style.use('seaborn-whitegrid')
import numpy as np 
```

```python
x = np.linspace(0,10,30)
y = np.sin(x)

plt.plot(x,y,'-o', color = 'red')
```

```python
plt.plot(x, y, '-p', color='blue'),
linewidth = 3, 
markersize = 16,
markerfacecolor = 'cyan',
markeredgecolor = 'green',
markeredgewidth = 2)

plt.ylim(-1.2, 1.2) 
```

```python
plt.scatter(x,y, marker='o', c='yellow', s=1000, edgecolors='black')
```
